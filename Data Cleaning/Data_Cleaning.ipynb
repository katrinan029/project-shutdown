```python
#Loading the csv file into pandas DataFrame 
import pandas as pd 
df = pd.read_csv("RBLSF.csv") 
df.head()
```

```python
# Get column names
column_names = df.columns
print(column_names)
```

```python
#drop columns that you recognize as unnecessary 
df.drop(["Location Id","Business Account Number","Ownership Name","Mail Address","Mail State", "Mail City","Mail Zipcode","NAICS Code","Parking Tax","Transient Occupancy Tax","LIC Code","LIC Code Description","Supervisor District","Business Corridor","UniqueID","SF Find Neighborhoods","Current Police Districts","Current Supervisor Districts","Analysis Neighborhoods","Neighborhoods"], axis = 1, inplace = True) 
  
df
```

```python
#Get details of the data that is not based in San Francisco 
indexNames = df[ df['City'] != "San Francisco" ].index
# Delete these row indexes from dataFrame
df.drop(indexNames , inplace=True)
df
```

```python
#extract the DataFrame to an output csv file 
df.to_csv('extracted_data.csv', sep='\t', encoding='utf-8')
```
